```python
def scotts_bins(data):
    std_dev = np.std(data)
    n = len(data)
    return int((3.5 * std_dev) / (n ** (1/3)))

fig, axes = plt.subplots(nrows=len(numerical_cols.columns), ncols=1, figsize=(6, 4 * len(numerical_cols.columns)))

for i, col in enumerate(numerical_cols.columns):
    sns.histplot(numerical_cols[col], bins=32, ax=axes[i], kde=True)
    axes[i].set_title(f'Histogram for {col}')

plt.tight_layout()
```

```python
cols = ['Poster Path', 'Title',  'Budget Musd',
        'Revenue Musd', 'Vote Count', 'Vote Average', 'Popularity']
df_best = df[cols]
df_best.sample(3)
```

```python
df_best['Profit Musd'] = df['Revenue Musd'] - df['Budget Musd']
df_best['Return'] = df['Revenue Musd']/df['Budget Musd']
```

```python
df_best.set_index('Title', inplace=True)
```

```python
cols = ['Poster Path', 'Budget Musd']
subset = df_best[cols].iloc[:5]
subset
```

```python
filt = df_best['Budget Musd']>5
df_best.loc[filt, :].sort_values(by='Return', ascending=False)
```

```python
df_best['Budget Musd'].fillna(0, inplace=True)
df_best['Vote Average'].fillna(0, inplace=True)
```

```python
def sorting(n, by, ascending=False, min_bud=0, min_votes=0):
    filt = (df_best['Budget Musd'] >= min_bud) & (df_best['Vote Average']>=min_votes)
    df2 = df_best.loc[filt, ['Poster Path',by]].sort_values(by=by, ascending=ascending).head(n)

    return HTML(df2.to_html(escape=False))
```

```python
sorting(n=5, by='Budget Musd')
```

```python
sorting(n=5, by='Revenue Musd')
```

```python
sorting(n=5, by='Profit Musd')
```

```python
sorting(n=5, by='Profit Musd', ascending=True)
```

```python
sorting(n=5, by='Return', ascending=True, min_bud=0)
```

```python
mask_genres = df['Genres'].str.contains('Action') & df['Genres'].str.contains('Science Fiction')
mask_actor = df['Cast'].str.contains('Bruce Willis')
bruce = df.loc[mask_actor & mask_genres, ['Title', 'Poster Path', 'Vote Average']].sort_values(by='Vote Average', ascending=False)
HTML(bruce.to_html(escape=False))
```

```python
mask_director = df['Director'] == 'Quentin Tarantino'
mask_actor = df['Cast'].str.contains('Uma Thurman')
quentin = df.loc[mask_director&mask_actor, ['Title', 'Poster Path', 'Runtime']].sort_values(by='Runtime').set_index('Title')
HTML(quentin.to_html(escape=False))
```

```python
mask_studio = df['Production Companies'].str.contains('Pixar').fillna(False)
mask_time = df['Release Date'].between('2010-01-01', '2015-12-31')
pixar = df.loc[mask_studio&mask_time, ['Title', 'Poster Path', 'Revenue Musd', 'Release Date']].sort_values(by='Revenue Musd', ascending=False).set_index('Title').head(5)
HTML(pixar.to_html(escape=False))
```

```python
mask_genre = df['Genres'].str.contains('Action') | df['Genres'].str.contains('Thriller')
mask_lan = df['Original Language'] == 'en'
mask_vote_av = df['Vote Average'] >= 7.5
mask_vote_co = df['Vote Count'] >= 10
next_movie = df.loc[mask_genre&mask_lan&mask_vote_av&mask_vote_co, ['Title', 'Poster Path', 'Genres', 'Vote Average', 'Vote Count', 'Release Date']].sort_values(by='Release Date', ascending=False).set_index('Title').head(5)
HTML(next_movie.to_html(escape=False))
```

```python
df.set_index('Id', inplace=True)
```

```python
actors_df = df['Cast'].str.split('|', expand=True)
actors_df.head(5)
```

```python
actors_df = actors_df.stack().reset_index(level=1, drop=True).to_frame()
actors_df.head(5)
```

```python
actors_df.columns = ['Actor']
```

```python
actors_df = actors_df.merge(df[['Title', 'Revenue Musd', 'Vote Average', 'Popularity']], 
                            how='left', left_index=True, right_index=True)
actors_df.head(5)
```

```python
actors_df['Actor'].nunique()
```

```python
actors_df['Actor'].unique()
```

```python
plt.figure(figsize=(10, 4))
ax = sns.barplot(x=actors_df['Actor'].value_counts().head(20).index,
                 y=actors_df['Actor'].value_counts().head(20))

ax.set_title('Top 20 Active Actor')
ax.set_ylabel('Number of Movies')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha="right")
plt.show()
```
